### Visualizing Data with Histograms

Import Libraries and Load the Dataset

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r'D:\New folder (4)\Coursera\DA Capstone\cleaned_survey_data3.csv'

dfo = pd.read_csv(file_path)

dfo.head()    # Display the first five rows
```

```python
df = dfo.copy()
```

##### Visualizing the distribution of data

**Histogram of `CompTotal` (Total Compensation)**

Plot a histogram of `CompTotal` to visualize the distribution of respondents' total compensation.

```python
# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['CompTotal'], bins=50, kde=True, color='teal')

plt.title('Distribution of Total Compensation (Log Scaled)')
plt.xlabel('Log(CompTotal + 1)')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Drop missing or zero compensation values
comp_data = df['CompTotal'].dropna()


# Optional: Apply log transformation if distribution is highly skewed
comp_data_log = np.log1p(comp_data)

# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(comp_data_log, bins=50, kde=True, color='teal')

plt.title('Distribution of Total Compensation (Log Scaled)')
plt.xlabel('Log(CompTotal + 1)')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()
```

**Histogram of `YearsCodePro` (Years of Professional Coding Experience)**

Plot a histogram of `YearsCodePro` to analyze the distribution of coding experience among respondents.

```python
# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['YearsCodePro'], bins=30, kde=True, color='purple')

plt.title('Distribution of Professional Coding Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()
```

##### Visualizing the Composition of Data

**Histogram of Most Desired Databases (`DatabaseWantToWorkWith`)**

Visualize the most desired databases for future learning using a histogram of the top 5 databases.

```python
# Get the top 5 databases by count

top_dbs = df['DatabaseWantToWorkWith'].value_counts().nlargest(5)

print(top_dbs)
```

```python
# plot the histogram

plt.figure(figsize=(8,5))
sns.barplot(x=top_dbs.index, y=top_dbs.values, hue=top_dbs.index, dodge=False, legend=False)
plt.title('Top 5 Most Desired Databases to Learn Next Year')
plt.ylabel('Number of Respondents')
plt.xlabel('Database')
plt.tight_layout()
plt.show()
```

##### Histogram of Preferred Work Locations (RemoteWork)

**Histogram of Preferred Work Locations (`RemoteWork`)**

Use a histogram to explore the distribution of preferred work arrangements (remote work).

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='RemoteWork', hue='RemoteWork', dodge=False, legend=False)
plt.title('Distribution of Preferred Work Locations (RemoteWork)')
plt.xlabel('Preferred Work Location')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)  # Rotate if labels overlap
plt.show()
```

##### Visualizing Comparison of Data

**Histogram of Median `CompTotal` for Ages 45 to 60**

Plot the histogram for `CompTotal` within the age group 45 to 60 to analyze compensation distribution among mid-career respondents.

```python
# Filter data for age 45 to 60
age_filtered = df[(df['AgeNumeric'] >= 45) & (df['AgeNumeric'] <= 60)]

plt.figure(figsize=(10,6))
plt.hist(age_filtered['CompTotal'].dropna(), bins=30, edgecolor='black')
plt.title('Distribution of Compensation (CompTotal) for Ages 45 to 60')
plt.xlabel('CompTotal (Annual Compensation)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Histogram of Job Satisfaction (`JobSat`) by `YearsCodePro`**

Plot the histogram for `JobSat` scores based on respondents' years of professional coding experience.

```python
# Create bins for YearsCodeProNumeric
bins = [0, 2, 5, 10, 20, 50, 60]
labels = ['0-2', '3-5', '6-10', '11-20', '21-50', '51+']
df['ExperienceGroup'] = pd.cut(df['YearsCodePro'], bins=bins, labels=labels, right=False)


# Plot histogram for JobSat grouped by ExperienceGroup using seaborn FacetGrid
g = sns.FacetGrid(df, col='ExperienceGroup', col_wrap=3, height=4, sharex=True, sharey=True)
g.map(plt.hist, 'JobSat', bins=10, edgecolor='black')
g.set_axis_labels('Job Satisfaction (JobSat)', 'Count')
g.set_titles('YearsCodePro: {col_name} years')
plt.tight_layout()
plt.show()
```
